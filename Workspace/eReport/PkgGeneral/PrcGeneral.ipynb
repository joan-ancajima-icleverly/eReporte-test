```
import pyodbc
import pandas as pd

class PrcGeneral:

    def __init__(self, driver='', server='', port='', database='', username='', password='', id_solicitud=''):

        self.server = server
        self.port = port
        self.database = database
        self.username = username
        self.password = password
        self.driver = driver
        
        self.id_solicitud = id_solicitud

        # Por defecto nos conectaremos a Azure SQL - Auqui
        self.server = 'auqui.database.windows.net'
        self.port = '1433'
        self.database = 'dbnormativo'
        self.username = 'administrador'
        self.password = 'Auqui$2020'
        self.driver = '{ODBC Driver 17 for SQL Server}'         


    def databaseConnect(self):                 

        connection = pyodbc.connect('DRIVER='+self.driver+';SERVER='+self.server+';PORT='+self.port+';DATABASE='+self.database+';UID='+self.username+';PWD='+ self.password)

        return connection

        connection.close()
        
        
    def obtenerParametros(self):

      #Abriendo la conexion
      connection = PrcGeneral().databaseConnect()

      #Obteniendo parametros
      select_Solicitud = ("SELECT ID, FECHA, EMPRESA, ARCHIVO_ENTRADA, TABLA_DESTINO, TABLA_DESTINO_2 FROM SOLICITUD WHERE ID = " + str(self.id_solicitud))
      sql_Solicitud = pd.read_sql_query(select_Solicitud, connection)
      df_Solicitud = pd.DataFrame(sql_Solicitud)

      id_solicitud = df_Solicitud.loc[0, 'ID']
      fecha_proceso = df_Solicitud.loc[0, 'FECHA']
      empresa = df_Solicitud.loc[0, 'EMPRESA']
      archivo_entrada = df_Solicitud.loc[0, 'ARCHIVO_ENTRADA']
      tabla_destino = df_Solicitud.loc[0, 'TABLA_DESTINO']
      tabla_destino_2 = df_Solicitud.loc[0, 'TABLA_DESTINO_2']

      return {
        'id_solicitud':id_solicitud, 
        'fecha_proceso':fecha_proceso,
        'empresa':empresa,
        'archivo_entrada':archivo_entrada,
        'tabla_destino':tabla_destino,
        'tabla_destino_2':tabla_destino_2}
    
    def cambioDiario(self, fechaProceso, tipoCambio):
      
      #Abriendo la conexion
      connection = PrcGeneral().databaseConnect()

      select_cambioDiario = ("SELECT MONTO FROM CAMBIO_DIARIO WHERE FECHA ='" + fechaProceso + "' AND TIPO_CAMBIO = '"+ tipoCambio +"'")
      sql_select_cambioDiario = pd.read_sql_query(select_cambioDiario, connection)
      df_select_cambioDiario = pd.DataFrame(sql_select_cambioDiario)
      cambio_diario = df_select_cambioDiario.loc[0,'MONTO']

      return cambio_diario

```
