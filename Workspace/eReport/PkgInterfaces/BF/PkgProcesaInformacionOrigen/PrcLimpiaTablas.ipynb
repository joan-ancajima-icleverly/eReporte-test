```
import pandas as pd
from datetime import datetime
from datetime import timedelta
import numpy as np

class PrcLimpiaTablas():
  
  def __init__(self, id_solicitud):
      self.id_solicitud = id_solicitud
      
      #Conexion
      connection = PrcGeneral().databaseConnect()
      self.connection = connection
      
      cursor = self.connection.cursor()
      self.cursor = cursor
  
  def limpiarTablas(self):
      
      #Limpiando la tabla
      delete = """DELETE FROM CLIENTE_PRODUCTO_CONCEPTO;
                  DELETE FROM CLIENTE_PRODUCTO;
                  DELETE FROM CLIENTE;
                  DELETE FROM CLIENTE_DEUDOR_PRODUCTO_DETALLE;
                  DELETE FROM CLIENTE_DEUDOR_PRODUCTO;
                  DELETE FROM CLIENTE_DEUDOR_DETALLE
                  DELETE FROM CLIENTE_DEUDOR"""
      self.cursor.execute(delete)
      self.connection.commit()
      
      self.cursor.close() 
            
#Ejecucion prueba
objeto = PrcLimpiaTablas(4)
objeto.limpiarTablas()
```
